## Venn diagram examples.

A collection of the examples from the [Python graph gallery](https://python-graph-gallery.com/)

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import numpy as np
 
```

```python
# There are multiple ways to create a venn diagram with two circles.

# Here 10 is in group A, 5 in group B, and the overlap is 2
venn2(subsets = (10, 5, 2), set_labels = ('Group A', 'Group B'))
plt.show()
 
```

```python
# A second way is to define the sets explicitly. Set A comes first, then set B

venn2([set(['A', 'F', 'C', 'D']), set(['D', 'E', 'F'])])
plt.show()

```

```python
# We can make a diagram with three groups, too!

from matplotlib_venn import venn3
 
# Specify the number in each group, then the pairwise overlaps, then the tri-overlap
venn3(subsets = (10, 8, 22, 6,9,4,2))
plt.show()
```

```python
# You can even get very fancy!

from matplotlib_venn import venn3_circles
 
# Define the venn diagram here
v = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels = ('A', 'B', 'C'))
 

v.get_patch_by_id('100').set_alpha(1.0)
v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('100').set_text('Unknown')
v.get_label_by_id('A').set_text('Set "A"')
c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')
c[0].set_lw(1.0)
c[0].set_ls('dotted')
 

plt.title("Sample Venn diagram")
plt.annotate('Unknown set', xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))
 
    
plt.show()
```
